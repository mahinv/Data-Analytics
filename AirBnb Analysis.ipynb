```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(file,encoding='unicode_escape', low_memory=False)
```

```python
df['host_since'] = pd.to_datetime(df['host_since'])
```

```python
df.info()
```

```python
paris_listing = df.query('city == "Paris"').loc[:,['host_since', 'neighbourhood', 'city', 'accommodates', 'price']]
```

```python
paris_listing = paris_listing.dropna()
```

```python
paris_listing.head(10)
```

```python
paris_listing.describe()
```

```python
paris_listing.query('price == 0 and accommodates != 0')
```

```python
paris_listing_neighbourhood = paris_listing.groupby('neighbourhood').mean('price').sort_values('price')


paris_listing_neighbourhood['price'].tail()
    
```

```python
paris_listing_accommodation = paris_listing.query("neighbourhood == 'Elysee'").groupby('accommodates').mean('price').sort_values(by = 'price').reset_index()
paris_listing_accommodation
```

```python
paris_listing_accommodation_1 = paris_listing.query("neighbourhood == 'Elysee'").groupby('accommodates')[('neighbourhood')].count().reset_index()
paris_listing_accommodation_1
```

```python
new_paris = paris_listing_accommodation_1.merge(paris_listing_accommodation, on = 'accommodates')
```

```python
new_paris.head()
```

```python
paris_listing_accommodation = paris_listing.query("neighbourhood == 'Elysee'").groupby('accommodates').agg({'price':'mean'}).sort_values(by = 'price')
```

```python
paris_listing_accommodation
```

```python
paris_listing_over_time = paris_listing.set_index('host_since').resample('y').agg({'neighbourhood':'count','price':'mean'})
```

```python
paris_listing_over_time
```

```python
paris_listing_neighbourhood
```

```python
import seaborn as sns
(paris_listing_neighbourhood.plot.barh(title = 'Average listing price by Paris neighbourhood', xlabel = 'Price per night',ylabel = 'Neighbourhood',legend = None))
sns.despine()
```

```python
(paris_listing_accommodation.plot.barh(title = 'Average listing price by Accommodates', xlabel = 'Price per night',ylabel = 'Accommodation Capacity',legend = None))
sns.despine()
```

```python
paris_listing_over_time['neighbourhood'].plot(ylabel = 'New Hosts', title = 'New AirBnB in Paris over time')
sns.despine()
```

```python
paris_listing_over_time['price'].plot(ylabel = 'Prices', title = 'Avg Price in Paris over time')
sns.despine()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()

ax.plot(paris_listing_over_time.index,paris_listing_over_time['neighbourhood'],label='New Hosts',c = 'Blue')

ax.set_ylabel ('Host count')

ax2 = ax.twinx()

ax2.plot(paris_listing_over_time.index,paris_listing_over_time['price'],label='Avg Price', c='Pink')
ax2.set_ylim(0)
ax2.set_ylabel ('Avg Price')


ax.set_title ('2015 Regulations lead to fewer hosts, higher prices')
```
